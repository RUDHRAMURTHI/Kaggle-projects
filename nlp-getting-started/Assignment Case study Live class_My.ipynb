```python
# Filtering only the customer tweets.
r1=[]
r2=[]
r3=[]
r4=[]
for t in results1:
    if t.author.screen_name !='IndiGo6E':
        r1.append(t.full_text)
        print(t.full_text)
for t in results2:
    if t.author.screen_name !='lufthansa':
        r2.append(t.full_text)
        print(t.full_text)
for t in results3:
    if t.author.screen_name !='MAS':
        r3.append(t.full_text)
        print(t.full_text) 
for t in results4:
    if t.author.screen_name !='emirates':
        r4.append(t.full_text)
        print(t.full_text)
```

```python
#Combining all the tweets.
master_results=[]
for text in r1:
    master_results.append(text)
for text in r2:
    master_results.append(text)
for text in r3:
    master_results.append(text)
for text in r4:
    master_results.append(text)    
```

```python
# Have to remove retweet, https, #, multiple commas, have to remove @
cleaned_Airlines_tweets= []
for tweet in master_results:
    # Removing retweet
    tweet = re.sub(r"RT @[\w]*:","", tweet)
    # Removing https
    tweet = re.sub(r"https?://[A-Za-z0-9./]*","", tweet)
    # Removing multiple commas
    tweet = re.sub(r",+|\s,+",", ", tweet)
    #removing #, @
    tweet = re.sub(r"[@#]","", tweet)
    # handling the new lines \n
    tweet = re.sub(r"\n+|\s\n+"," ", tweet)
    cleaned_Airlines_tweets.append(tweet)


```

```python
Parsed_Airline_tweets = []
patterns = """
P: {<NN.?>*<VB.?>*}
NP: {<DT>?<RB.?>*<JJ.?><NN.?>?}
"""
chunkparser = nltk.RegexpParser(patterns)
for tweet in cleaned_Airlines_tweets:
    tweet = word_tokenize(tweet)
    tweet = pos_tag(tweet)
    Parsed_Airline_tweets.append(chunkparser.parse(tweet))
```

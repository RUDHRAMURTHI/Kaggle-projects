```python
def Cabclass(x):
    if x[0] == 'A':
        return('A')
    elif x[0] == 'B':
        return('B')
    elif x[0] == 'C':
        return('C')
    elif x[0] == 'D':
        return('D')
    elif x[0] == 'E':
        return('E')
    elif x[0] == 'F':
        return('F')
    elif x[0] == 'G':
        return('G')
    elif x[0] == 'T':
        return('T')
    else :
        return(np.nan)
```
